```python


# Crear dos dataframe y luego unirlos en uno solo
data1 = {'nombre': ['Maria', 'Carolina'],
         'edad': [25, 30]}
df1 = pd.DataFrame(data1)

data2 = {'nombre': ['Pedro', 'Sebastian'],
         'edad': [32, 25]}
df2 = pd.DataFrame(data2)

# Concatenacion vertial (por filas)
vertical_concatenated = pd.concat([df1, df2], ignore_index=True)
print("Vertically concatenated DataFrame:")
print(vertical_concatenated)
```

```python
# Creacion de dos data frames con columnas adicioanles
data3 = {'nombre': ['Eva', 'Frank'],
         'Salario': [500000, 650000]}
df3 = pd.DataFrame(data3)

data4 = {'nombre': ['laura', 'Henry'],
         'Salario': [550030, 755000]}
df4 = pd.DataFrame(data4)

# Concatenate horizontally (along columns)
horizontal_concatenated = pd.concat([df3, df4], axis=1)
print("\nHorizontally concatenated DataFrame:")
print(horizontal_concatenated)
```

```python
# Una muestra a prtir de dos distribuciones diferentes normal
means = 500, 200
stdevs = 45, 10
dist = pd.DataFrame(
       np.random.normal(loc=means, scale=stdevs, size=(1000, 2)),
       columns=['ingresPhD', 'ingresmaster'])
dist.agg(['min', 'max', 'mean', 'std']).round(decimals=2)
```

```python
fig, ax = plt.subplots()
dist.plot.kde(ax=ax, legend=False, title='Histogram: A vs. B')
dist.plot.hist(density=True, ax=ax)
ax.set_ylabel('Probabilidad')
ax.grid(axis='y')
ax.set_facecolor('#d8dcd6')
```
